```python
from haystack.document_stores.in_memory import InMemoryDocumentStore
document_store = InMemoryDocumentStore()
```

```python
# Update the datasets library to the latest version
#!pip install -U datasets
from datasets import load_dataset
from haystack import Document
dataset=load_dataset("bilgeyucel/seven-wonders",split='train')
```

```python
dataset
```

```python
#for doc in dataset:
  #print(doc['content'])
```

```python
for doc in dataset:
  doc = Document(content=doc['content'],meta=doc['meta'])

```

```python
doc
```

```python
from haystack.components.embedders.sentence_transformers_document_embedder import SentenceTransformersDocumentEmbedder

# Instantiate the class
doc_embeder = SentenceTransformersDocumentEmbedder(model="sentence-transformers/all-MiniLM-L6-v2")
doc_embeder.warm_up() #to start llm
```

```python
doc_with_emb = doc_embeder.run([doc])
document_store.write_documents(doc_with_emb["documents"])
```

```python
#embeddings for user querry
from haystack.components.embedders import SentenceTransformersDocumentEmbedder

# Instantiate the class
embedder = SentenceTransformersDocumentEmbedder(model="sentence-transformers/all-MiniLM-L6-v2")
```

```python
from haystack.components.retrievers.in_memory import InMemoryEmbeddingRetriever
retriever = InMemoryEmbeddingRetriever(document_store=document_store)
```

```python
from haystack.components.builders import PromptBuilder

template = """
Given the following information, answer the question.

Context:
{% for document in documents %}
{{ document.content }}
{% endfor %}

Question: {{question}}
Answer:
"""

prompt_builder = PromptBuilder(template=template)
```
